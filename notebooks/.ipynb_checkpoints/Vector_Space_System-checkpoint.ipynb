```python
import pandas as pd

df = pd.read_csv("../data/metadata.csv")
print(df)
```

#### Preparing the dataset
We decided just to work with the papers of the PDF_JSON corpus. Therefore, the first step is to delete from the dataframe the elements that are not in that folder. The number of examples is reduced from 192509 to 79755. Still, there are more documents in the pdf_json than in the dataframe (over 84000) which means that some documents in the corpus are repeated or are not mapped with a cord_uid (or both). For our model, we are going to ignore the documents that do not have a cord_uid (o podemos usarlo todo, esto lo vamos viendo).

```python
df = df[df.pdf_json_files.notnull()]
df = df.reset_index(drop = True)
print(df)
```

Next, we drop the columns that will not add information to our information retrieval system (such as the licenses or the doi) and that do not help to map each example of the dataframe with a document in the pdf_json corpus.

```python
columns_to_delete = ["doi", "source_x", "pmcid", "pubmed_id", "license", "mag_id", "who_covidence_id", "arxiv_id", "pmc_json_files", "url", "s2_id"]
df_original = df
df = df_original.drop(columns_to_delete, axis = 1)
print(df)
```
